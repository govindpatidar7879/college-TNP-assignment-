**For Loop problem**

**1)WAP using a for loop to print all Armstrong numbers between 100 and 999.**

```python
print("Armstrong numbers between 100 and 999 are:")
for num in range(100, 1000):
    d1 = num // 100        
    d2 = (num // 10) % 10  
    d3 = num % 10  
    if (d1**3 + d2**3 + d3**3) == num:
        print(num)
```

**2)WAP to generate and display the first n prime numbers using a for loop.**

```python
n = int(input("Enter VALUE OF n"))
count = 0
num = 2
for num in range(2, 10000):  
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
        count += 1
    if count == n:
        break
```

**4)WPA using a for loop to print a pyramid of stars (*) of height n.**

```python
n = int(input("Enter pyramid height: "))
for i in range(1, n+1):
    for j in range(n-i):
        print(" ", end="")
    for k in range(2*i-1):
        print("*", end="")
    print()
```
